```python
import numpy as np
from matplotlib import pyplot as plt
import mpl_style as mplt
plt.style.use(['mvstyle', 'one_piece'])
```

```python
IC_tt = np.loadtxt('IceCube/IC_tautau.dat', unpack=True)
IC_bb = np.loadtxt('IceCube/IC_bb.dat', unpack=True)
IC_WW = np.loadtxt('IceCube/IC_WW.dat', unpack=True)
IC_PICO2L = np.loadtxt('IceCube/IC_PICO2L.dat', unpack=True)

PICO_interp = mplt.make_spline(IC_PICO2L[0], IC_PICO2L[1], 0.001, 3, xscale = 'log', yscale = 'log')
PICO_x = np.logspace(np.log10(IC_PICO2L[0][0]), np.log10(IC_PICO2L[0][-1]), 100)
PICO_y = PICO_interp(PICO_x)
```

```python
fig, ax = plt.subplots()

mplt.loglog(ax, IC_bb[0], IC_bb[1], label = r'$b\bar{b}$', linewidth = 2, c = 'C00')
mplt.loglog(ax, IC_WW[0], IC_WW[1], label = r'$W^+W^-$', linewidth=2, c = 'C01')
mplt.loglog(ax, IC_tt[0], IC_tt[1], label = r'$\tau^+\tau^-$', linewidth=2, c = 'C02')
mplt.loglog(ax, PICO_x, PICO_y, c = 'C03')
plt.text(10, 5e-39, r'\textsf{PICO-2L}', color='C03', rotation=-70)
plt.text(7, 5e-39, r'\textsf{(2015)}', color='C03', rotation=-70)


plt.xlim(IC_PICO2L[0][0], 1e4)
plt.ylim(1e-41, 1e-37)
plt.xlabel(r'$m_\chi\;(\mathrm{GeV})$')
plt.ylabel(r'$\sigma_{\chi p}^{\mathrm{SD}}\;(\mathrm{cm}^{-2})$')
plt.legend(loc = 'lower left')
plt.title(r'$\mathrm{IceCube\;(2011-2015)}$')
plt.savefig('plots/IceCube_2016.pdf')
```
