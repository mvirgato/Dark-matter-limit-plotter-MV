```python
import numpy as np
from matplotlib import pyplot as plt
import matplotlib as mpl
from glob import glob
from scipy.interpolate import interp1d, BSpline, splrep

import mpl_style as mplt

plt.style.use(['mvstyle', 'one_piece'])
```

```python
fig = plt.figure(figsize=(7, 8/3))
ax = fig.add_subplot(111)

ax.spines[['right', 'top', 'left']].set_visible(False)
ax.tick_params(which='major',direction='out', top = False, left = False, right = False)
ax.tick_params(which='minor',direction='out', top = False, left = False, right = False)
ax.get_yaxis().set_visible(False)


plt.xscale('log')
plt.xlim(1e-30, 1e24)
plt.xlabel(r'$\mathrm{Dark\; Matter\; Mass }\;(\mathrm{GeV}/c^2)$')

plt.ylim(0, 8/3)

locmajx = mpl.ticker.LogLocator(base = 10.0, numticks=20)
ax.xaxis.set_major_locator(locmajx)
ax.xaxis.set_major_formatter(mplt.MyLogFormatter())


dy = 0.1

# WIMPs
col = 'C00'
ybase = 0.1
ytop = ybase + 0.2
masses = [1e-2, 1e5]
xmid = 10**((np.log10(masses[0]) + np.log10(masses[1]))/2.0)
plt.fill_between(masses,[ybase, ybase], y2 = ytop, color = col, alpha = 0.5, edgecolor=col)
plt.text(xmid, ytop+0.05, r'$\mathrm{WIMPs}$', ha='center', va = 'bottom', color = col, fontsize = 12)

# Composites
col = 'C10'
ybase = 0.1
ytop = ybase + 0.2
masses = [1e6, 1e24]
xmid = 10**((np.log10(masses[0]) + np.log10(masses[1]))/2.0)
plt.fill_between(masses,[ybase, ybase], y2 = ytop, color = col, alpha = 0.5, edgecolor=col)
plt.text(xmid, ytop+0.05, r'$\mathrm{Composite/MACROs}$', ha='center', va = 'bottom', color = col, fontsize = 12)


# WIMPZILA
col = 'C02'
ybase = 0.7
ytop = ybase + 0.2
masses = [1e11, 1e16]
xmid = 10**((np.log10(masses[0]) + np.log10(masses[1]))/2.0)
plt.fill_between(masses,[ybase, ybase], y2 = ytop, color = col, alpha = 0.5, edgecolor=col)
plt.text(xmid, ytop+0.05, r'$\mathrm{WIMPzilla}$', ha='center', va = 'bottom', color = col, fontsize = 12)

# QBalls
col = 'C06'
ybase = 1.3
ytop = ybase + 0.2
masses = [1e8, 1e13]
xmid = 10**((np.log10(masses[0]) + np.log10(masses[1]))/2.0)
plt.fill_between(masses,[ybase, ybase], y2 = ytop, color = col, alpha = 0.5, edgecolor=col)
plt.text(xmid, ytop+0.05, r'$\mathrm{Q-balls}$', ha='center', va = 'bottom', color = col, fontsize = 12)

# Dark Photons
col = 'C04'
ybase = 0.1
ytop = ybase + 0.2
masses = [1e-30, 1e-5]
xmid = 10**((np.log10(masses[0]) + np.log10(masses[1]))/2.0)
plt.fill_between(masses,[ybase, ybase], y2 = ytop, color = col, alpha = 0.5, edgecolor=col)
plt.text(xmid, ytop+0.05, r'$\mathrm{Dark\;Photons}$', ha='center', va = 'bottom', color = col, fontsize = 12)

# ALPs
col = 'C05'
ybase = 0.7
ytop = ybase + 0.2
masses = [1e-30, 10**-5.5]
xmid = 10**((np.log10(masses[0]) + np.log10(masses[1]))/2.0)
plt.fill_between(masses,[ybase, ybase], y2 = ytop, color = col, alpha = 0.5, edgecolor=col)
plt.text(xmid, ytop+0.05, r'$\mathrm{Axion\;Like\;Particles}$', ha='center', va = 'bottom', color = col, fontsize = 12)

# QCD Axion
col = 'C03'
ybase = 1.3
ytop = ybase + 0.2
masses = [1e-20, 10**-11]
xmid = 10**((np.log10(masses[0]) + np.log10(masses[1]))/2.0)
plt.fill_between(masses,[ybase, ybase], y2 = ytop, color = col, alpha = 0.5, edgecolor=col)
plt.text(xmid, ytop+0.05, r'$\mathrm{QCD\;Axion}$', ha='center', va = 'bottom', color = col, fontsize = 12)


# Sterile neutrinos
col = 'C07'
ybase = 1.9
ytop = ybase + 0.2
masses = [1e-7, 10**-3]
xmid = 10**((np.log10(masses[0]) + np.log10(masses[1]))/2.0)
plt.fill_between(masses,[ybase, ybase], y2 = ytop, color = col, alpha = 0.5, edgecolor=col)
plt.text(xmid, ytop+0.05, r'$\mathrm{Sterile\;}\nu$', ha='center', va = 'bottom', color = col, fontsize = 12)

# Hidden Sector DM
col = 'C08'
ybase = 0.7
ytop = ybase + 0.2
masses = [10**-5.05, 10**6]
xmid = 10**((np.log10(masses[0]) + np.log10(masses[1]))/2.0)
plt.fill_between(masses,[ybase, ybase], y2 = ytop, color = col, alpha = 0.5, edgecolor=col)
plt.text(xmid, ytop+0.05, r'$\mathrm{Hidden\;Sector\;DM}$', ha='center', va = 'bottom', color = col, fontsize = 12)


# Ultralight Scalars
col = 'C09'
ybase = 1.9
ytop = ybase + 0.2
masses = [1e-30, 1e-9]
xmid = 10**((np.log10(masses[0]) + np.log10(masses[1]))/2.0)
plt.fill_between(masses,[ybase, ybase], y2 = ytop, color = col, alpha = 0.5, edgecolor=col)
plt.text(xmid, ytop+0.05, r'$\mathrm{Ultralight\;Scalars}$', ha='center', va = 'bottom', color = col, fontsize = 12)

# Gravitino
col = 'C01'
ybase = 1.3
ytop = ybase + 0.2
masses = [1e-6, 1e4]
xmid = 10**((np.log10(masses[0]) + np.log10(masses[1]))/2.0)
plt.fill_between(masses,[ybase, ybase], y2 = ytop, color = col, alpha = 0.5, edgecolor=col)
plt.text(xmid, ytop+0.05, r'$\mathrm{Gravitino}$', ha='center', va = 'bottom', color = col, fontsize = 12)


plt.savefig('DM_model_landscape.pdf')
```
